```python
## Without latex installation
%%manim -v WARNING -qm SimpleShapes
from manim import *

class SimpleShapes(Scene):
    def construct(self):
        # Test without text first
        circle = Circle(radius=1.5, color=BLUE, fill_opacity=0.3)
        square = Square(side_length=2, color=RED, fill_opacity=0.3)
        triangle = Triangle(color=GREEN, fill_opacity=0.3).scale(1.5)

        # Arrange shapes
        circle.shift(LEFT * 3)
        square.shift(RIGHT * 3)

        self.play(Create(circle))
        self.play(Create(square))
        self.play(Create(triangle))
        self.wait(1)

        # Animate movements
        self.play(circle.animate.shift(UP))
        self.play(square.animate.shift(DOWN))
        self.play(triangle.animate.rotate(PI/2))
        self.wait(1)
```

```python
##With Latex Installation
%%manim -v WARNING -qm TextTest
from manim import *

class TextTest(Scene):
    def construct(self):
        # Test simple text
        text1 = Text("Hello Manim!", font_size=48)
        self.play(Write(text1))
        self.wait(1)

        # Test mathematical expression
        equation = MathTex(r"\int_a^b f(x)dx = F(b) - F(a)")
        equation.next_to(text1, DOWN, buff=1)
        self.play(Write(equation))
        self.wait(1)

        # Test matrix
        matrix = Matrix([[1, 2], [3, 4]])
        matrix.next_to(equation, DOWN, buff=1)
        self.play(Create(matrix))
        self.wait(2)
```

```python
%%manim -v WARNING -qm --format=mp4 ListSlicingIntro
from manim import *

class ListSlicingIntro(Scene):
    def construct(self):
        # Config
        self.camera.background_color = "#1e1e1e"
        title_font = 48
        content_font = 36
        code_font = 28

        # Title (3 seconds)
        title = Text("Python List Slicing", font_size=title_font, color=BLUE)
        subtitle = Text("The Ultimate Guide", font_size=content_font, color=YELLOW)
        subtitle.next_to(title, DOWN, buff=0.3)

        self.play(Write(title), run_time=1)
        self.play(Write(subtitle), run_time=0.8)
        self.wait(1.2)

        # Clear title
        self.play(FadeOut(title), FadeOut(subtitle), run_time=0.5)

        # Sample list (5 seconds)
        list_title = Text("Sample List:", font_size=content_font, color=WHITE)
        list_title.to_edge(UP, buff=0.5)

        fruits = ["apple", "banana", "cherry", "date", "elderberry"]
        list_display = Text(str(fruits), font_size=code_font, color=GREEN)
        list_display.next_to(list_title, DOWN, buff=0.5)

        indices = Text("Index:   0        1        2        3        4",
                      font_size=code_font-4, color=YELLOW)
        indices.next_to(list_display, DOWN, buff=0.3)

        neg_indices = Text("Neg Index: -5      -4      -3      -2      -1",
                          font_size=code_font-4, color=ORANGE)
        neg_indices.next_to(indices, DOWN, buff=0.2)

        self.play(Write(list_title), run_time=0.5)
        self.play(Write(list_display), run_time=0.8)
        self.play(Write(indices), run_time=0.7)
        self.play(Write(neg_indices), run_time=0.7)
        self.wait(2.3)

        # Clear and show basic syntax (6 seconds)
        self.play(
            FadeOut(list_title),
            FadeOut(list_display),
            FadeOut(indices),
            FadeOut(neg_indices),
            run_time=0.5
        )

        syntax_title = Text("Basic Syntax:", font_size=content_font, color=WHITE)
        syntax_title.to_edge(UP, buff=0.5)

        syntax_formula = Text("list[start : stop : step]",
                            font_size=content_font, color=GREEN)
        syntax_formula.next_to(syntax_title, DOWN, buff=0.5)

        explanation = Text("start: inclusive, stop: exclusive",
                         font_size=content_font-4, color=YELLOW)
        explanation.next_to(syntax_formula, DOWN, buff=0.3)

        self.play(Write(syntax_title), run_time=0.5)
        self.play(Write(syntax_formula), run_time=0.8)
        self.play(Write(explanation), run_time=0.7)
        self.wait(4)

        # Clear and show examples (8 seconds)
        self.play(
            FadeOut(syntax_title),
            FadeOut(syntax_formula),
            FadeOut(explanation),
            run_time=0.5
        )

        examples_title = Text("Common Examples:", font_size=content_font, color=WHITE)
        examples_title.to_edge(UP, buff=0.5)

        # Example 1
        ex1_code = Text('fruits[1:4] â†’ ["banana", "cherry", "date"]',
                       font_size=code_font, color=GREEN)
        ex1_code.next_to(examples_title, DOWN, buff=0.5)

        # Example 2
        ex2_code = Text('fruits[:3] â†’ ["apple", "banana", "cherry"]',
                       font_size=code_font, color=BLUE)
        ex2_code.next_to(ex1_code, DOWN, buff=0.3)

        # Example 3
        ex3_code = Text('fruits[::2] â†’ ["apple", "cherry", "elderberry"]',
                       font_size=code_font, color=ORANGE)
        ex3_code.next_to(ex2_code, DOWN, buff=0.3)

        self.play(Write(examples_title), run_time=0.5)
        self.play(Write(ex1_code), run_time=0.7)
        self.wait(0.5)
        self.play(Write(ex2_code), run_time=0.7)
        self.wait(0.5)
        self.play(Write(ex3_code), run_time=0.7)
        self.wait(5.1)

        # Clear and show advanced slicing (8 seconds)
        self.play(
            FadeOut(examples_title),
            FadeOut(ex1_code),
            FadeOut(ex2_code),
            FadeOut(ex3_code),
            run_time=0.5
        )

        advanced_title = Text("Advanced Slicing:", font_size=content_font, color=WHITE)
        advanced_title.to_edge(UP, buff=0.5)

        # Negative indices
        neg_ex = Text('fruits[-3:-1] â†’ ["cherry", "date"]',
                     font_size=code_font, color=GREEN)
        neg_ex.next_to(advanced_title, DOWN, buff=0.5)

        # Reverse
        rev_ex = Text('fruits[::-1] â†’ reversed list',
                     font_size=code_font, color=BLUE)
        rev_ex.next_to(neg_ex, DOWN, buff=0.3)

        # Step
        step_ex = Text('fruits[1:5:2] â†’ ["banana", "date"]',
                      font_size=code_font, color=ORANGE)
        step_ex.next_to(rev_ex, DOWN, buff=0.3)

        self.play(Write(advanced_title), run_time=0.5)
        self.play(Write(neg_ex), run_time=0.7)
        self.wait(0.3)
        self.play(Write(rev_ex), run_time=0.7)
        self.wait(0.3)
        self.play(Write(step_ex), run_time=0.7)
        self.wait(5)

        # Final summary (5 seconds)
        self.play(
            FadeOut(advanced_title),
            FadeOut(neg_ex),
            FadeOut(rev_ex),
            FadeOut(step_ex),
            run_time=0.5
        )

        summary_title = Text("Key Takeaways:", font_size=content_font, color=WHITE)
        summary_title.to_edge(UP, buff=0.8)

        takeaway1 = Text("â€¢ [start:stop] - selects range",
                        font_size=code_font, color=GREEN)
        takeaway1.next_to(summary_title, DOWN, buff=0.4, aligned_edge=LEFT)

        takeaway2 = Text("â€¢ [:] - copies entire list",
                        font_size=code_font, color=BLUE)
        takeaway2.next_to(takeaway1, DOWN, buff=0.2, aligned_edge=LEFT)

        takeaway3 = Text("â€¢ [::-1] - reverses list",
                        font_size=code_font, color=ORANGE)
        takeaway3.next_to(takeaway2, DOWN, buff=0.2, aligned_edge=LEFT)

        final_msg = Text("Master slicing in 30 seconds! ğŸ¯",
                       font_size=content_font-4, color=YELLOW)
        final_msg.next_to(takeaway3, DOWN, buff=0.5)

        self.play(Write(summary_title), run_time=0.5)
        self.play(Write(takeaway1), run_time=0.5)
        self.play(Write(takeaway2), run_time=0.5)
        self.play(Write(takeaway3), run_time=0.5)
        self.play(Write(final_msg), run_time=0.5)
        self.wait(2.5)
```
