```python
# Visualizing linearity and non-linearity in input space
# IMPORTANT: Each plot is its own figure, no explicit colors used.

import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# ---------- 1. Linearly separable dataset ----------
X1 = np.random.randn(50, 2) + np.array([2, 2])
X2 = np.random.randn(50, 2) + np.array([-2, -2])

plt.figure()
plt.scatter(X1[:, 0], X1[:, 1])
plt.scatter(X2[:, 0], X2[:, 1])
x = np.linspace(-6, 6, 100)
plt.plot(x, -x)  # linear decision boundary
plt.title("Linearly Separable Dataset (Linear Boundary)")
plt.show()


# ---------- 2. Linearly NON-separable dataset (XOR) ----------
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])

plt.figure()
for label in np.unique(y):
    plt.scatter(X[y==label,0], X[y==label,1])
plt.title("Linearly Non-Separable Dataset (XOR)")
plt.show()


# ---------- 3. Non-linear but separable dataset (Circle) ----------
theta = np.linspace(0, 2*np.pi, 100)
inner_x = 0.5 * np.cos(theta)
inner_y = 0.5 * np.sin(theta)

outer_x = 1.5 * np.cos(theta)
outer_y = 1.5 * np.sin(theta)

plt.figure()
plt.scatter(inner_x, inner_y)
plt.scatter(outer_x, outer_y)
plt.title("Non-Linear Separable Dataset (Circular Boundary)")
plt.show()


# ---------- 4. Non-linear and NON-separable dataset ----------
X1 = np.random.randn(100, 2)
X2 = np.random.randn(100, 2)

plt.figure()
plt.scatter(X1[:,0], X1[:,1])
plt.scatter(X2[:,0], X2[:,1])
plt.title("Non-Linear Non-Separable Dataset (Overlapping Classes)")
plt.show()
```
